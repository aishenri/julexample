```julia
using ODE,Plots
pyplot(size=(700,300),leg=false,
guidefont=font(9), titlefont=font(9));
oscillator(t, y) = [y[2], - 3* + y[1] - y[2] / 10]
initial = [1.0,0.0]; 
t = 0:0.01:50   
t,y = ode23(oscillator, initial, t)
y=hcat(y...).';
y[1:5,:];
v=y[:,1];
v=y[:,2];
o=plot(t,y,title="Mouvement",
bg=RGB(.2,.2,.2),
xlabel ="Temps",ylabel = "Vitesse")
l=plot(v,v,title="portrait de phase",
bg=RGB(.2,.2,.2),
xlabel ="Temps",ylabel = "Vitesse")
plot(o,l,layout=2)
```
